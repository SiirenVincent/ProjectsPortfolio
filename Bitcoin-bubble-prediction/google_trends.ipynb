```python
#import libraries to use
from pytrends.request import TrendReq
import pandas as pd
import time
```

```python

startTime = time.time()
pytrend = TrendReq(hl='en-US', tz=360)


colnames = ["keywords"] #from trends.csv dataset
df = pd.read_csv("trends.csv", names=colnames) #read dataset to get keywords to be used
df2 = df["keywords"].values.tolist()
df2.remove("Keywords")

def year(date):
  dataset = []
  for x in range(0,len(df2)):
     keywords = [df2[x]]
     pytrend.build_payload(
     kw_list=keywords,
     cat=0,
     timeframe=date,
     geo='')

```

```python
#get data from 2012-01-01 to 2012-09-01
year('2012-01-01 2012-09-01')
dataset = []
data = pytrend.interest_over_time()
if not data.empty:
          data = data.drop(labels=['isPartial'],axis='columns')
          dataset.append(data)
result = pd.concat(dataset, axis=1)
result.reset_index()
```
